````python
# prompt: plot the cumulative variace of pca for all the possibel pronviopal components
## to do
import matplotlib.pyplot as plt
import numpy as np
pca = PCA()
pca.fit(numeric_data)
plt.plot(np.cumsum(pca.explained_variance_ratio_), marker='o')
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
plt.show()
````

````python
pca = PCA(n_components=2)
pca.fit(numeric_data)
pca_data = pca.transform(numeric_data)

# Create a DataFrame with the principal components and target labels
pca_df = pd.DataFrame({
    "pca_1": pca_data[:, 0],
    "pca_2": pca_data[:, 1],
    "target": data["target"]
})

# Visualize the PCA results with a scatter plot
sns.scatterplot(x="pca_1", y="pca_2", hue="target", data=pca_df)
plt.title("PCA Visualization of Heart Disease Data")
plt.show()
````

````python
from sklearn.manifold import TSNE

# Initialize and fit the TSNE model
tsne = TSNE(n_components=2)
tsne_data = tsne.fit_transform(numeric_data)

# Create a DataFrame with the TSNE components and target labels
tsne_df = pd.DataFrame({
    "tsne_1": tsne_data[:, 0],
    "tsne_2": tsne_data[:, 1],
    "target": data["target"]
})

# Visualize the TSNE results with a scatter plot
sns.scatterplot(x="tsne_1", y="tsne_2", hue="target", data=tsne_df)
plt.title("TSNE Visualization of Heart Disease Data")
plt.show()
````

1. Percentage of Samples with Disease:
To find the percentage of samples with the disease, we need to know the total number of samples and how many of them have the disease.
Once we have that information, we can calculate the percentage using the formula: [ \text{Percentage with Disease} = \frac{\text{Number of Samples with Disease}}{\text{Total Number of Samples}} \times 100]
2. Continuous Features Showing Significant Differences:
To identify features that significantly differ between presence and absence of the disease, we can perform statistical tests (e.g., t-tests, ANOVA) or use non-parametric methods (e.g., Mann-Whitney U test, Kruskal-Wallis test).
The three features showing significant differences would be those with the lowest p-values from these tests.
Keep in mind that significance doesn’t always imply practical importance. We might need to consider effect sizes as well.
3. PCA and t-SNE Separation:
Principal Component Analysis (PCA) and t-Distributed Stochastic Neighbor Embedding (t-SNE) are dimensionality reduction techniques.
While PCA aims to maximize variance, t-SNE focuses on preserving pairwise similarities.
Whether we see clear separation depends on the data and the inherent structure:
If the disease-related features have distinct clusters, both PCA and t-SNE might reveal separation.
If the data is inherently noisy or overlapping, separation might not be clear.
4. Optimal Number of Principal Components:
Determining the optimal number of principal components involves looking at the explained variance.
We can plot the cumulative explained variance against the number of components and choose a point where adding more components doesn’t significantly increase explained variance.
Techniques like scree plots or the “elbow method” can help decide the optimal number.
5. Highly Correlated Continuous Features:
To find the most correlated features, we can compute the correlation matrix (e.g., Pearson correlation) for all continuous features.
The pairs with high absolute correlation coefficients (close to 1 or -1) are the most correlated.
Keep an eye out for multicollinearity—when features are highly correlated, it can affect model interpretability.

````python
##  one hot encoding of categorical features in data
def onehot_encode(df, columns, prefixes):
    df = df.copy()
    for column, prefix in zip(columns, prefixes):
        dummies = pd.get_dummies(df[column], prefix=prefix)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(column, axis=1)
    return df
````
